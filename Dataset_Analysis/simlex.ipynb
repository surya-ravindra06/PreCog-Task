```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Simlex Data into dataframe
df = pd.read_csv('../SimLex-999/SimLex-999.txt', sep='\t')
print(df.head())
```

```python
# Get all the simlex scores and convert them to a numpy array
simlex_scores = df['SimLex999'].values
# print(simlex_scores)
```

```python
# Plot a graph of the simlex scores to see the distribution: histogram
plt.hist(simlex_scores, bins=20, color='blue', edgecolor='black', linewidth=1.2)
plt.title('Histogram of Simlex Scores')
plt.xlabel('Simlex Score')
plt.ylabel('Frequency')
plt.show()
```
